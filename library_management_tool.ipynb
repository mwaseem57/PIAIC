```python
books = [
    {"bookId":"1", "title": "To Kill a Mockingbird", "author": "Harper", "year": 1949, "status":"Available"},
    {"bookId":"2", "title": "1984", "author": "George Orwell", "year": 1960, "status":"Checked Out"},
    {"bookId":"3", "title": "The Great Gatsby" , "author": "F. Scott Fitzgerald", "year": 1925, "status":"Available"}
    ]
users = [
    {"userId":"1", "name":"waseem", "borrowedBook":[]},
    {"userId":"2", "name":"fahad", "borrowedBook":[]},
    {"userId":"3", "name": "Johny", "borrowedBook":[]}
    ]

# View All Book
def viewBooks():

    for b in books:
        print(f"{b['title']}\n")

# Add a New Book
def addBook(books):
    book = {
        'bookId': input("Enter Book ID: "),
        'title': input("Enter Title: "),
        'author': input("Enter Author: "),
        'year': input("Enter Year of Publication: "),
        'status': input("Enter Availability Status: ")
    }
    books.append(book)
    print(books)

# Search Book
def searchBook():
    search = input("Enter book title or author: ").lower()
    found = False

    for book in books:
        if search in book["title"].lower() or search in book["author"].lower():
            print(f"Found: {book['title']} by {book['author']}, Status: {book['status']}")
            found = True

    if not found:
        print("The book doesn't exist.")

#Borrow Book
def borrowBook():
    uId = input("Enter your userId: ")
    userFound = False
    bookFound = False
    for user in users:
        if uId == user['userId']:
            userFound = True
            if userFound:
                search = input("Enter the title of the book you want to borrow: ").lower()
                for book in books:
                    if search in book['title'].lower() and book['status']=="Available":
                        bookFound = True
                        user['borrowedBook'].append(book['title'])
                        book['status'] ="Checked Out"
                        print(f"The Book {search} has been borrowed")
                        return
        
    if not userFound:
        print(f"User ID {uId} not found.")
        return 
    if not bookFound:
        print(f"The book {search} is not available or doesn't exist")

#Return Book
def returnBook():
    uId = input("Enter your userId: ")
    returningUserFound = False
    returningBookFound = False
    for user in users:
        if uId == user['userId']:
            returningUserFound = True
        else:
            break
        returningBook = input("Please enter the book title you want to return: ").lower()
        for book in user['borrowedBook']:
            if returningBook in book.lower():
                returningBookFound = True
                user['borrowedBook'].remove(book)
                print(user['borrowedBook'])
                print(f"The book {book} has been returned successfully")
                return

    if not returningUserFound:
        print("User ID not found.")
        return
    if not returningBookFound:
        print("The book has been already returned or doesn't exist")


# View Users
def viewUsers():
    for u in users:
        print(f"{u['name']}\n")

# Main function
def main():
    while True:
        print("\nPlease choose an option:")
        print("1. View all books")
        print("2. Add a book")
        print("3. Search for a book")
        print("4. Borrow a book")
        print("5. Return a book")
        print("6. View all users")
        print("7. Exit")

        choice = input("Enter your choice (1-7): ")
        print(f"Enter your Choice: {choice}")

        if choice == "1":
            viewBooks()
        elif choice == "2":
            addBook(books)
        elif choice == "3":
            searchBook()
        elif choice == "4":
            borrowBook()
        elif choice == "5":
            returnBook()
        elif choice == "6":
            viewUsers()
        elif choice == "7":
            print("Exiting the library system")
            break
        else:
            print("Invalid choice. Please select a valid option.")
        print("----------------------------------------")

if __name__ == "__main__":
    print("Welcome to the Community Library System!")
    print("----------------------------------------")

    main() 

```
